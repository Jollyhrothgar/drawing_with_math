# Introduction

The [Collatz Conjecture](https://en.wikipedia.org/wiki/Collatz_conjecture) states that if you start with any positive integer $N$, one can apply the following rules:

$$
N_{n+1} = {N \over 2}, if\ n\ is\ even
$$
$$
N_{n+1} = 3 N + 1,\ if\ n\ is\ odd
$$

The conjecture states that no matter what $N$ you start with, you will always reach $1$.

```python
def collatz(N, arr=None):
  if N == 0:
    raise ValueError("ZERO IS NOT ALLOWED!")
  if arr is None:
    arr = []
  if N == 1:
    arr.append(N)
    return arr
  elif N % 2 == 0:
    arr.append(N)
    N /= 2
    return collatz(int(N), arr)
  elif N % 2 == 1:
    arr.append(N)
    N = 3*N + 1
    return collatz(int(N), arr)
```

```python
for i in range(1,10):
  print(collatz(i))
```

```python
fig, ax = plt.subplots()
for i in range(1,100):
  plt.plot(collatz(i))
plt.xlabel("Sequence Position")
plt.ylabel("Sequence Value")
plt.title("Collatz Sequences")
```

```python
res = []
for i in range(1, 100000):
  seq = collatz(i)
  
  res.append(
    {
      'starting_number': i,
      'length':len(seq),
    }
  )
df = pd.DataFrame(res)
```
